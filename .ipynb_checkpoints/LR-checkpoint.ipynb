```python
from paddle.utils.plot import Ploter
import paddle
import paddle.fluid as fluid
import numpy as np
import math
import sys
import pandas as pd
import matplotlib.pyplot as plt
```

训练数据提供器,每次读入BATCH_SIZE大小的数据

```python
SAVE_DIRNAME = 'LR_Adam_model'
df = pd.read_csv('./data/BABAprocessed.csv')
```

```python
df.info()
```

```python
#z-score
#normalization =lambda x: (x-np.min(x))/(np.std(x))
#df_normalization = df[['Open','High','Low','Close','Adj Close','Volume','rate','label']].apply(normalization)

#01
normalization = lambda x: (x-np.mean(x))/(np.max(x)-np.min(x))
df_normalization = df[['Open','High','Low','Close','Adj Close','Volume','rate','label']].apply(normalization)

#部分特征进行归一化
#df_new = df.copy().drop(['Open','High','Low','Close','Adj Close','Volume','rate'],axis=1)
#df_new = pd.concat([df_new,df_normalization],axis=1)

df_normalization
```

```python
data = np.array(df_normalization)
data
```

```python
ratio = 0.8 # 训练集和验证集的划分比例
offset = int(data.shape[0]*ratio)
train_data = data[:offset]
test_data = data[offset:]
# 每次数据的条数
BATCH_SIZE = 256
BUF_SIZE = 1000
def reader_creator(train_data):  
    # 生产一行数据
    def reader():  
        for d in train_data:  
            yield d[:-1], d[-1:]  
    return reader

train_reader = paddle.batch(
    paddle.reader.shuffle(
        reader_creator(train_data), buf_size=BUF_SIZE),
        batch_size=BATCH_SIZE)

test_reader = paddle.batch(
    paddle.reader.shuffle(
        reader_creator(test_data), buf_size=BUF_SIZE),
        batch_size=BATCH_SIZE)
```

```python
# 定义输入层的形状和数据类型
x = fluid.data(name='x', shape=[None,7], dtype='float32') 
# 定义输出层的形状和数据类型
y = fluid.data(name='y', shape=[None, 1], dtype='float32') 
# 输出层,连接输入和输出的全连接层,size为输出单元的数目，act激活函数
y_predict = fluid.layers.fc(input=x, size=1, act=None) 
# 获取默认/全局主函数,用于保存layer的operator和veriable
main_program = fluid.default_main_program()
# 获取默认/全局启动程序
startup_program = fluid.default_startup_program() 
# 平方差损失函数,但获得的只是一个batch的,利用标签数据和输出的预测数据估计方差
cost = fluid.layers.square_error_cost(input=y_predict, label=y) 
# 对方差求均值，得到平均损失
avg_loss = fluid.layers.mean(cost) 
```

```python
# 克隆main_program得到test_program,用于预测数据
# 有些operator在训练和测试之间的操作是不同的，例如batch_norm，使用参数for_test来区分该程序是用来训练还是用来测试
# 该api不会删除任何操作符,要在backward和optimization之前使用,
test_program = main_program.clone(for_test=True)
# 定义优化方法,使用随机梯度下降
#sgd_optimizer = fluid.optimizer.SGD(learning_rate=0.001)
#sgd_optimizer.minimize(avg_loss)
adam_optimizer = fluid.optimizer.AdamOptimizer(0.001)
adam_optimizer.minimize(avg_loss)
```

```python
 # 指明executor的执行场所
use_cuda = False
place = fluid.CUDAPlace(0) if use_cuda else fluid.CPUPlace()

###executor可以接受传入的program，并根据feed map(输入映射表)和fetch list(结果获取表)向program中添加数据输入算子和结果获取算子。使用close()关闭该executor，调用run(...)执行program。
exe = fluid.Executor(place)
```

# 创建训练过程
训练需要有一个训练程序和一些必要参数，并构建了一个获取训练过程中测试误差的函数。必要参数有executor,program,reader,feeder,fetch_list，executor表示之前创建的执行器，program表示执行器所执行的program，是之前创建的program，如果该项参数没有给定的话则默认使用default_main_program，reader表示读取到的数据，feeder表示前向输入的变量，fetch_list表示用户想得到的变量或者命名的结果。

```python
num_epochs = 100

def train_test(executor, program, reader, feeder, fetch_list):
    accumulated = 1 * [0]
    count = 0
    for data_test in reader():
        outs = executor.run(program=program,
                            feed=feeder.feed(data_test),
                            fetch_list=fetch_list)
        accumulated = [x_c[0] + x_c[1][0] for x_c in zip(accumulated, outs)] # 累加测试过程中的损失值
        count += 1 # 累加测试集中的样本数量
    return [x_d / count for x_d in accumulated] # 计算平均损失
```

# 训练主循环

```python
%matplotlib inline

# 定义输入数据维度
feeder = fluid.DataFeeder(place=place, feed_list=[x, y])
exe.run(startup_program)
train_prompt = "train cost"
test_prompt = "test cost"
plot_prompt = Ploter(train_prompt, test_prompt)
step = 0

exe_test = fluid.Executor(place)
```

# 设置监听器
打印每一次训练后损失的变化以及相应的输出

```python
for pass_id in range(num_epochs):
    for data_train in train_reader():
        avg_loss_value, = exe.run(main_program,
                                  feed=feeder.feed(data_train),
                                  fetch_list=[avg_loss])
        # 每10个批次记录并输出一下训练损失
        if step % 10 == 0: 
            plot_prompt.append(train_prompt, step, avg_loss_value[0])
            plot_prompt.plot()
            print("%s, Step %d, Cost %f" %
                  (train_prompt, step, avg_loss_value[0]))
         # 每100批次记录并输出一下测试损失
        if step % 100 == 0:  
            # 使用测试执行器,之前clone的program,数据由test_reader提供,每组数据为设定的batch_size=20
            test_metics = train_test(executor=exe_test,
                                     program=test_program,
                                     reader=test_reader,
                                     fetch_list=[avg_loss],
                                     feeder=feeder)
            plot_prompt.append(test_prompt, step, test_metics[0])
            plot_prompt.plot()
            print("%s, Step %d, Cost %f" %
                 (test_prompt, step, test_metics[0]))
            # 如果准确率达到要求，则停止训练


        step += 1

        if math.isnan(float(avg_loss_value[0])):
            sys.exit("got NaN loss, training failed.")
plot_prompt.plot('./image/LR_sgd_loss.png')
#保存训练参数到之前给定的路径中
if SAVE_DIRNAME is not None:
    fluid.io.save_inference_model(SAVE_DIRNAME, ['x'], [y_predict], exe)
```

# 预测

```python
infer_exe = fluid.Executor(place)
#  创建作用域
inference_scope = fluid.core.Scope()
```

```python
# 保存图片
def save_result(points1, points2):
    import matplotlib
    matplotlib.use('Agg')
    x1 = [idx for idx in range(len(points1))]
    y1 = points1
    y2 = points2
    l1 = plt.plot(x1, y1, 'r-', label='predictions')
    l2 = plt.plot(x1, y2, 'g-', label='true')
    plt.plot(x1, y1, 'r-', x1, y2, 'g-')
    plt.title('Adam predictions VS true')
    plt.legend()
    plt.savefig('./image/LR_adam_results.png')
```

```python
# 将所有的参数赋值给作用域
with fluid.scope_guard(inference_scope):
    [inference_program, feed_target_names,
     fetch_targets] = fluid.io.load_inference_model(SAVE_DIRNAME, infer_exe) # 载入预训练模型
    batch_size = 266

    infer_reader = paddle.batch(
        paddle.reader.shuffle(
        reader_creator(train_data), buf_size=BATCH_SIZE * 20), batch_size=batch_size) # 准备测试集

    infer_data = next(infer_reader())
    infer_feat = np.array(
        [data[0] for data in infer_data]).astype("float32") # 提取测试集中的数据
    infer_label = np.array(
        [data[1] for data in infer_data]).astype("float32") # 提取测试集中的标签

    assert feed_target_names[0] == 'x'
    results = infer_exe.run(inference_program,
                            feed={feed_target_names[0]: np.array(infer_feat)},
                            fetch_list=fetch_targets) # 进行预测

    # 取出二维数组里的值并生成一个list用于后续的升序排列
    results_li = []
    for idx, val in enumerate(results[0]):
        results_li.append(val[0])
    infer_label_li = []
    for idx, val in enumerate(infer_label):
        infer_label_li.append(val[0])
    results_np = np.array(results_li)
    order = results_np.argsort(axis=0)
    results_np = results_np[order]
    infer_label_np = np.array(infer_label_li)
    infer_label_np = infer_label_np[order]



    
    save_result(results_np, infer_label_np)
```

```python
save_result(results_np, infer_label_np)
```
