```python
import paddle
import paddle.fluid as fluid
import numpy as np
import math
import sys
import pandas as pd
import matplotlib.pyplot as plt
from paddle.utils.plot import Ploter
```

```python
# model存储位置
SAVE_DIRNAME = 'LSTM_Adam_model'
df = pd.read_csv('./data/BABAprocessed.csv')
df.info()
```

normal_01 = lambda x: (x-np.mean(x))/(np.max(x)-np.min(x))

z_score = lambda x: (x-np.min(x))/(np.std(x))
## 2 需要归一化的数据

### 2.2 除目标外的所有特征值

```python
#z-score
#normalization =lambda x: (x-np.min(x))/(np.std(x))
#df_normalization = df[['Open','High','Low','Close','Adj Close','Volume','rate','label']].apply(normalization)

#01
normalization = lambda x: (x-np.mean(x))/(np.max(x)-np.min(x))
df_normalization = df[['Open','High','Low','Close','Adj Close','Volume','rate','label']].apply(normalization)

#部分特征进行归一化
#df_new = df.copy().drop(['Open','High','Low','Close','Adj Close','Volume','rate'],axis=1)
#df_new = pd.concat([df_new,df_normalization],axis=1)

df_normalization
```

```python
data = np.array(df_normalization,dtype='float32')
print('数据类型：',type(data))
print('数据个数：', data.shape[0])
print('数据形状：', data.shape)
print('数据第一行：', data[0])
```

# 划分数据集以及设置产生器

```python
# 训练集和验证集的划分比例
ratio = 0.8 
offset = int(data.shape[0]*ratio)
train_data = data[:offset]
test_data = data[offset:]
# 每次数据的条数
BATCH_SIZE = 32

```

```python
# reader
def reader_creator(train_data):  
    # 生产一行数据
    def reader():  
        for d in train_data:  
            yield d[:-1], d[-1:]  
    return reader

train_reader = paddle.batch(
        reader_creator(train_data), 
        batch_size=BATCH_SIZE)

test_reader = paddle.batch(
        reader_creator(test_data),
        batch_size=BATCH_SIZE * 10)
```

# 搭建lstm网络

```python
DIM = 1
hid_dim2 = 1
x = fluid.layers.data(name='x', shape=[DIM], dtype='float32', lod_level=1)
y = fluid.layers.data(name='y', shape=[1], dtype='float32')

#lstm,fc为全连接层
fc0 = fluid.layers.fc(input=x, size=DIM * 4)
#h为隐藏状态，c为单元状态，即记忆细胞
lstm_h, c = fluid.layers.dynamic_lstm(input=fc0, size=DIM * 4, is_reverse=False)
```

```python
lstm_max = fluid.layers.sequence_pool(input=lstm_h, pool_type='max')
# 激活函数
lstm_max_tanh = fluid.layers.tanh(lstm_max)
# 全连接层
prediction = fluid.layers.fc(input=lstm_max_tanh, size=hid_dim2, act='tanh')

```

```python
main_program = fluid.default_main_program()
# 获取默认/全局启动程序
startup_program = fluid.default_startup_program() 
# 平方差损失函数,但获得的只是一个batch的,利用标签数据和输出的预测数据估计方差
loss = fluid.layers.square_error_cost(input=prediction, label=y)
avg_loss = fluid.layers.mean(x=loss)
```

# 设置优化器
## 两种优化方法，sgd与adam

```python
# 克隆main_program得到test_program,用于预测数据
# 有些operator在训练和测试之间的操作是不同的，例如batch_norm，使用参数for_test来区分该程序是用来训练还是用来测试
# 该api不会删除任何操作符,要在backward和optimization之前使用,
test_program = main_program.clone(for_test=True)
# 定义优化方法,使用随机梯度下降或adam
#sgd_optimizer = fluid.optimizer.SGD(learning_rate=0.001)
#sgd_optimizer.minimize(avg_loss)
adam_optimizer = fluid.optimizer.AdamOptimizer(0.001)
adam_optimizer.minimize(avg_loss)
```

```python
 # 指明executor的执行场所
use_cuda = False
place = fluid.CUDAPlace(0) if use_cuda else fluid.CPUPlace()

###executor可以接受传入的program，并根据feed map(输入映射表)和fetch list(结果获取表)向program中添加数据输入算子和结果获取算子。使用close()关闭该executor，调用run(...)执行program。
exe = fluid.Executor(place)
```

# 创建训练过程

```python
num_epochs = 100

def train_test(executor, program, reader, feeder, fetch_list):
    accumulated = 1 * [0]
    count = 0
    for data_test in reader():
        outs = executor.run(program=program,
                            feed=feeder.feed(data_test),
                            fetch_list=fetch_list)
        accumulated = [x_c[0] + x_c[1][0] for x_c in zip(accumulated, outs)] # 累加测试过程中的损失值
        count += 1 # 累加测试集中的样本数量
    return [x_d / count for x_d in accumulated] # 计算平均损失
```

# 训练主循环

```python
%matplotlib inline
# 定义输入数据维度
feeder = fluid.DataFeeder(place=place, feed_list=[x, y])
exe.run(startup_program)
train_prompt = "train cost"
test_prompt = "test cost"
plot_prompt = Ploter(train_prompt, test_prompt)
step = 0
exe_test = fluid.Executor(place)
```

# 设置监听器

```python

for pass_id in range(num_epochs):
    for data_train in train_reader():
        avg_loss_value, = exe.run(main_program,
                                  feed=feeder.feed(data_train),
                                  fetch_list=[avg_loss])
        # 每10个批次记录并输出一下训练损失
        if step % 10 == 0: 
            plot_prompt.append(train_prompt, step, avg_loss_value[0])
            plot_prompt.plot()
            print("%s, Step %d, Cost %f" %
                  (train_prompt, step, avg_loss_value[0]))
         # 每100批次记录并输出一下测试损失
        if step % 100 == 0:  
            # 使用测试执行器,之前clone的program,数据由test_reader提供
            test_metics = train_test(executor=exe_test,
                                     program=test_program,
                                     reader=test_reader,
                                     fetch_list=[avg_loss],
                                     feeder= feeder)
            plot_prompt.append(test_prompt, step, test_metics[0])
            plot_prompt.plot()
            print("%s, Step %d, Cost %f" %
                 (test_prompt, step, test_metics[0]))
            


        step += 1

        if math.isnan(float(avg_loss_value[0])):
            sys.exit("got NaN loss, training failed.")
plot_prompt.plot('./image/LSTMs_Adam_loss.png')
#保存训练参数到之前给定的路径中
plot_prompt
if SAVE_DIRNAME is not None:
    fluid.io.save_inference_model(SAVE_DIRNAME, ['x'], [prediction], exe)
```

```python
infer_exe = fluid.Executor(place)
#  创建作用域
inference_scope = fluid.core.Scope()
```

```python
# 保存图片
def save_result(points1, points2):
    import matplotlib
    matplotlib.use('Agg')
    x1 = [idx for idx in range(len(points1))]
    y1 = points1
    y2 = points2
    l1 = plt.plot(x1, y1, 'r-', label='predictions')
    l2 = plt.plot(x1, y2, 'g-', label='true')
    plt.plot(x1, y1, 'r-', x1, y2, 'g-')
    plt.title('LSTM predictions VS true')
    plt.legend()
    plt.savefig('./image/LSTMs_Adam_results.png')
```

```python
def convert2LODTensor(temp_arr, len_list):
    temp_arr = np.array(temp_arr) 
    temp_arr = temp_arr.flatten().reshape((-1, 1))
    print(temp_arr.shape)
    return fluid.create_lod_tensor(
        data=temp_arr,
        recursive_seq_lens =[len_list],
        place=fluid.CPUPlace()
        )
    

def get_tensor_label(mini_batch):  
    tensor = None
    labels = []
    
    temp_arr = []
    len_list = []
    for _ in mini_batch:  
        labels.append(_[1]) 
        temp_arr.append(_[0]) 
        len_list.append(len(_[0])) 
    tensor = convert2LODTensor(temp_arr, len_list)    
    return tensor, labels

my_tensor = None
labels = None





for mini_batch in test_reader():
    my_tensor,labels = get_tensor_label(mini_batch)#其实就是变成tensor格式的x和y
    break
```

```python
# 将所有的参数赋值给作用域
with fluid.scope_guard(inference_scope):
    # 载入预训练模型 
    [inference_program, feed_target_names,
     fetch_targets] = fluid.io.load_inference_model(SAVE_DIRNAME, infer_exe) 
    # 进行预测
    results = exe.run(inference_program,
                      feed= {'x': my_tensor}, #{feed_target_names[0]:my_tensor },和上面保存模型时统一
                      fetch_list=fetch_targets)
result_print = results[0].flatten()
save_result(labels, result_print)
```
